### Part 1: Data preparation
Create a data preparation and cleaning function that does the following:
- Has a single input that is a file name string
- Reads data (the data is comma separated, has a row header and the first column `EventID` is the index) into a pandas `dataframe`
- Cleans the data
  - Convert the feature `Label` to numeric (choose the minority class to be equal to 1)
    - Create a feature `Y` with numeric label
    - Drop the feature `Label`
  - If a feature has missing values (i.e., `-999`): 
    - Create a dummy variable for the missing value
      - Call the variable `orig_var_name` + `_mv` where `orig_var_name` is the name of the actual var with a missing value
      - Give this new variable a 1 if the original variable is missing
    - Replace the missing value with the average of the feature (make sure to compute the mean on records where the value isn't missing). You may find pandas' `.replace()` function useful.
- After the above is done, rescales the data so that each feature has zero mean and unit variance (hint: look up sklearn.preprocessing)
- Returns the cleaned and rescaled dataset

Hint: as a guide, this function can easily be done in less than 15 lines.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

def cleanBosonData(infile_name):
    # set first column as index
    df = pd.read_csv(infile_name, index_col=0)
    # sort label and take the least populated class
    minority_class = df.Label.value_counts().index[-1]
    # minority class is 1 everything else is 0
    Y = df.Label.apply(lambda x: 1 if x == minority_class else 0)
    df = df.drop('Label', axis=1)
    # for each column replace -999 with NaN, replace NaNs with column mean
    df = df.apply(
        lambda feature: feature.replace(-999, np.nan).replace(np.nan, feature.mean(skipna=True)))
    #centering and scaling
    df = (df - df.mean()) / df.std(ddof=0)

    # ddof here refers to the degrees of freedom, which is number of samples,
    # that goes into denominator of the formula; pandas uses 1 by default
    # for variance calculation, numpy (and therefore sklearn) uses 0

    df['Y'] = Y

    # this can also be done using sklearn
    #df = df.apply(StandardScaler().fit_transform)
    return df
```

For each model above, plot the ROC curve of both models on the same plot. Make sure to use the test set for computing and plotting. In the legend, also print out the Area Under the ROC (AUC) for reference.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


# splitting into X (every column except the last one) and y
train_X, train_y = data_train.ix[:, :-1], data_train.Y
test_X, test_y = data_test.ix[:, :-1], data_test.Y

# fitting on training data
lr = LogisticRegression(C=1e30)
lr.fit(train_X, train_y)

# by default we are predicting binary values for classes, e.g. 0 or 1, but we can also
# predict probabilities of belonging to each class, e.g. [0.87 , 0.13].
# probabilities produce a smooth ROC curve
# LinearRegression can predict them without any extra commands, SVC needs
# probability=True

mode = 'probabilities'

if mode == 'probabilities':
    svm = SVC(kernel='linear', probability=True)
else:
    svm = SVC(kernel='linear')

svm.fit(train_X, train_y)


# metrics for auc and roc
from sklearn import metrics

import itertools
# colors as neverending iterator
colors = itertools.cycle(
    ['cyan', 'indigo', 'seagreen', 'yellow', 'blue', 'darkorange'])


for name, model in (('SVM', svm), ('LR', lr)):
    # predicting on test data
    if mode == 'probabilities':
        # it returns two columns, one for each class {0,1}, we need the 2nd,
        # for the 1s
        preds = model.predict_proba(test_X)[:, 1]
    else:
        preds = model.predict(test_X)
    fpr, tpr, t = metrics.roc_curve(test_y.values, preds)
    print(fpr, tpr, name)
    roc_auc = metrics.auc(fpr, tpr)
    color = next(colors)
    plt.plot(fpr, tpr, lw=2, color=color,
             label='ROC %s  (area = %0.2f)' % (name, roc_auc))

# baseline 45 degrees
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='k',
         label='45 degree line')
plt.xlabel('fpr')
plt.ylabel('tpr')
#plt.legend(loc="lower right")

#plot threshold
plt.plot(fpr, t, linestyle='--', lw=2, color='r',
         label='Threshold')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)

```

### Part 3: Model selection with cross-validation
We think we might be able to improve the performance of the SVM if we perform a grid search on the hyper-parameter $C$.  Because we only have 1000 instances, we will have to use cross-validation to find the optimal $C$.

1\. Write a cross-validation function that does the following:
- Takes as inputs a dataset, a label name, # of splits/folds (`k`), a sequence of values for $C$ (`cs`)
- Performs two loops
  - Outer Loop: `for each f in range(k)`:
    - Splits the data into `data_train` & `data_validate` according to cross-validation logic
  - Inner Loop: `for each c in cs`:
    - Trains an SVM on training split with `C=c, kernel="linear"`
    - Computes AUC_c_k on validation data
    - Stores AUC_c_k in a  dictionary of values
- Returns a dictionary, where each key-value pair is: `c:[auc-c1,auc-c2,..auc-ck]`

```python
# Code here
def xValSVM(dataset, label_name, k, cs):
    if k == 1:
        raise Exception('Number of folds should be more than one.')
    # randomizing
    dataset = dataset.sample(frac=1)

    # select non-label columns, provided we know only label column name
    x_columns = dataset.columns.difference(
        dataset.columns[dataset.columns == label_name])
    X = dataset[x_columns]
    y = dataset[label_name]
    n_samples = len(dataset)
    # an indices range, from zero
    indices = pd.Index(np.arange(n_samples))
    fold_size = n_samples / k
    print(fold_size, n_samples , k)
    aucs = dict()
    current = 0
    # we iterate k times, moving validation set window to the right each step
    # i.e.    0..199,    199..398   , 398..597, etc., and subtract (in set-theoretic sense) this from the full index at each step
    # getting 200..2000, 0..199 + 399..2000, 0..397 + 598..2000 etc., as a
    # training set

    for f in range(k):
        start, stop = current, current + fold_size

        # last one gets all that remains
        if f == k:
            validation_idx = indices[start:]
        else:
            validation_idx = indices[start:stop]

        training_idx = indices.difference(validation_idx)

        # applying indices
        train_X, train_y = X.iloc[training_idx], y.iloc[training_idx]
        validate_X, validate_y = X.iloc[validation_idx], y.iloc[validation_idx]
        # for each value C we train SVM on training set, predicting on validating set
        # calculate ROC and AUC, and save them in a list in a dictionary
        for c in cs:
            if not c in aucs:
                aucs[c] = []
            svm = SVC(kernel='linear', C=c)
            #svm = SVC(kernel='linear', C=c, probability=True)
            svm.fit(train_X, train_y)
            preds = svm.predict(validate_X)
            #preds = svm.predict_proba(validate_X)[:,1]
            fpr, tpr, _ = metrics.roc_curve(validate_y, preds)
            roc_auc = metrics.auc(fpr, tpr)
            aucs[c].append(roc_auc)
        current += fold_size

    return aucs
```
